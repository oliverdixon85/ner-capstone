## Do an initial training run

```python
from simpletransformers.ner import NERModel, NERArgs
import pandas as pd
from google.colab import files

model_args = NERArgs(learning_rate=2e-5, num_train_epochs=10, eval_batch_size=32, train_batch_size=32, weight_decay=1, overwrite_output_dir=True)
model_args.labels_list = ["B-LOC", "I-LOC", "B-ORG", "I-ORG", "B-DATE", "B-PER", "I-PER", "I-DATE", "O"]


model = NERModel(
    "xlmroberta", "xlm-roberta-base", 
    args=model_args,
)
```

## Upload and process training data

```python
from google.colab import files
uploaded = files.upload()
```

```python
def load_train():
  df_text = uploaded['train_swa.txt'].decode("utf-8")
  text = df_text.split('\n\n')

  df_train = pd.DataFrame()
  df_train['sentence_id'] = pd.Series()
  df_train['words'] = pd.Series()
  df_train['labels'] = pd.Series()

  sentence_num = []
  words = []
  tag = []
  for i in range(len(text)):
      sentence = text[i].split('\n')
      for j in range(len(sentence)-1):
          word = sentence[j].split(" ")
          sentence_num.append(i)
          words.append(word[0])
          tag.append(word[1])
     
  df_train['sentence_id'] = sentence_num
  df_train['words'] = words
  df_train['labels'] = tag
  return df_train
```

```python
df_train = load_train()
```

```python
model.train_model(df_train)
```

## Streamlit App 

```python
%%writefile app.py 
from simpletransformers.ner import NERModel, NERArgs
import streamlit as st 


st.title('')
    
activities = ['NER Checker']
choice = st.sidebar.selectbox("Select Activity",activities)
    
if choice == 'NER Checker':
  st.header('Named Entity Recognition')
  st.text("")
  st.text("")
  raw_text = st.text_area("Enter Text Here","Type Here")
  raw_text1 = raw_text.title()
  model_args = NERArgs(learning_rate=1e-5, num_train_epochs=1, eval_batch_size=8, train_batch_size=8, weight_decay=1, overwrite_output_dir=True)
  model_args.labels_list = ["B-LOC", "I-LOC", "B-ORG", "I-ORG", "B-DATE", "B-PER", "I-PER", "I-DATE", "O"]
  model = NERModel(
    "xlmroberta", "/content/outputs/", 
    args=model_args,
		)
  model.args.use_multiprocessing = False
    #	model = load_model_ner('Named Entity Recognition/best_model_state_ner_f2.bin')
        #model = load_model_ner('https://github.com/raviteja-ganta/deeplearning_streamlit_app/blob/main/Named%20Entity%20Recognition/best_model_state_ner_f2.bin')
if st.button("Analyze"):
  predictions, raw_outputs = model.predict([raw_text])
  text_dict = {}
  for i in range(len(predictions)):
    			res = {k: v for d in predictions[i] for k, v in d.items()}
    			text_dict.update(res)

  def entRecognizer(entDict, typeEnt):
    			entList = [ent for ent in entDict if entDict[ent] == typeEnt]
    			return entList

     #Using function to create lists of entities of each type
  entOrgA = entRecognizer(text_dict, "B-ORG")
  entOrgB = entRecognizer(text_dict, "B-ORG")
  entNone = entRecognizer(text_dict, "O")
  entPersonA = entRecognizer(text_dict, "B-PER")
  entPersonB = entRecognizer(text_dict, "I-PER")
  entDateA = entRecognizer(text_dict, "B-DATE")
  entDateB = entRecognizer(text_dict, "I-DATE")
  entLocationA = entRecognizer(text_dict, "B-LOC")
  entLocationB = entRecognizer(text_dict, "I-LOC")

      #Displaying entities of each type
  st.write("Organization Entities: " + str(entOrgA) + str(entOrgB))
  st.write("None Entities: " + str(entNone))
  st.write("Personal Entities: " + str(entPersonA) + str(entPersonB))
  st.write("Date Entities: " + str(entDateA) + str(entDateB))
  st.write("Location Entities: " + str(entLocationA) + str(entLocationB))

    
    
    
```
